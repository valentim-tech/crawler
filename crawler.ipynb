```python
# Função para extrair os dados da seção mt-9 e do JSON-LD correto
def extrair_dados(url, id):
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
        resposta = requests.get(url, headers=headers)
        resposta.raise_for_status()
        
        soup = BeautifulSoup(resposta.text, 'html.parser')
        
        # Dicionário para armazenar todos os dados
        dados = {'id': id, 'url': url}
        
        # 1. Extrair a seção mt-9 (tabela de características)
        secao = soup.find('div', class_='mt-9')
        if secao:
            linhas = secao.find_all('div', recursive=False)
            for linha in linhas:
                colunas = linha.find_all('div', recursive=False)
                if len(colunas) == 2:
                    atributo = colunas[0].get_text(strip=True)
                    valor = colunas[1].get_text(strip=True)
                    dados[atributo] = valor
        else:
            print(f"Seção 'mt-9' não encontrada em {url}")
        
        # 2. Extrair o JSON-LD correto
        # Buscar todos os scripts application/ld+json
        scripts = soup.find_all('script', type='application/ld+json')
        for script in scripts:
            try:
                json_data = json.loads(script.string)
                # Verificar se é o script do produto (@type: "Product")
                if json_data.get('@type') == 'Product':
                    # Adicionar os dados do JSON-LD ao dicionário
                    dados['nome'] = json_data.get('name', 'N/A')
                    dados['productID'] = json_data.get('productID', 'N/A')
                    dados['descricao'] = json_data.get('description', 'N/A')
                    dados['sku'] = json_data.get('sku', 'N/A')
                    dados['imagem'] = json_data.get('image', 'N/A')
                    dados['marca'] = json_data.get('brand', {}).get('name', 'N/A')
                    dados['categoria'] = json_data.get('category', 'N/A')
                    dados['condicao'] = json_data.get('itemCondition', 'N/A')
                    # Dados da oferta (offers)
                    if 'offers' in json_data and json_data['offers']:
                        oferta = json_data['offers'][0] if isinstance(json_data['offers'], list) else json_data['offers']
                        dados['preco'] = oferta.get('price', 'N/A')
                        dados['moeda'] = oferta.get('priceCurrency', 'N/A')
                        dados['disponibilidade'] = oferta.get('availability', 'N/A')
                    break  # Para após encontrar o script correto
            except json.JSONDecodeError:
                continue  # Ignora scripts inválidos
        
        return dados
    
    except Exception as e:
        print(f"Erro ao processar {url}: {e}")
        return None
```

```python
# Lista para armazenar os dados
dados_ofertas = []

# Processar os links com IDs sequenciais
for id, link in enumerate(links_ofertas, start=1):
    dados = extrair_dados(link, id)
    if dados:
        dados_ofertas.append(dados)
    time.sleep(1)  # Delay para evitar bloqueios

# Salvar em JSON
with open('caracteristicas_ofertas.json', 'w', encoding='utf-8') as arquivo_json:
    json.dump(dados_ofertas, arquivo_json, ensure_ascii=False, indent=4)

print("Dados salvos em 'caracteristicas_ofertas.json'")
```
